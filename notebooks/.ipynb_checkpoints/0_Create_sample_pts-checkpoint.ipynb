## Get random sample within polygon

```python
aoi_in = Path("../data/EParaguay.shp")
sampsize = 5000
subpoly=None
gs = get_sample_in_poly(aoi_in, sampsize, subpoly)

fig, ax = plt.subplots()
gdf = gpd.read_file(aoi_in)
gdf.plot(ax=ax, facecolor='none', edgecolor='steelblue')
gs.plot(ax=ax, color='r')
```

## Shift points to centroids of ref ras and make sample boxes

```python
ref_ras = Path("../data/samp_ras.tif")

gs_shift = move_points_to_pixel_centroids(gs, ref_ras, write_pts=False)

## Plot again to make sure that crs is correctly matched
fig, ax = plt.subplots()
gdf = gpd.read_file(aoi_in)
if (gdf.crs != gs_shift.crs):
    print("Reprojecting aoi file to match image crs")
    gdf = gdf.to_crs(gs_shift.crs)
gdf.plot(ax=ax, facecolor='none', edgecolor='steelblue')
gs_shift.plot(ax=ax, color='r')
```

```python
poly_file = "C:/Users/klobw/Desktop/Testing/polys_test8.shp"
res = 10
boxes = []
for index,row in gs_shift.iterrows():
    pt = (row.geometry.x, row.geometry.y)
    tl = (pt[0]-res/2, pt[1]+res/2)
    tr = (pt[0]+res/2, pt[1]+res/2)
    bl = (pt[0]-res/2, pt[1]-res/2)
    br = (pt[0]+res/2, pt[1]-res/2)
    box = [br,tr,tl,bl]
    boxes.append(box)
    
df = pd.DataFrame({'geometry': boxes})
df['geometry'] = df['geometry'].apply(shapely.geometry.Polygon)
df['FID'] = df.index
gdf = gpd.GeoDataFrame(df, crs=gs_shift.crs, geometry=df['geometry'])
gdf.to_file(poly_file, driver='ESRI Shapefile')

#plot_poly_on_index(1046, ref_ras, poly_file)
```
